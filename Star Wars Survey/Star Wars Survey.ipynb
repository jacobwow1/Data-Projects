# Cleaning and Exploring a *Star Wars* Survey Dataset

```python
import pandas as pd
star_wars = pd.read_csv('StarWars.csv', encoding='ISO-8859-1')
```

```python
star_wars.head(10)
```

```python
print(star_wars.columns)
```

## Cleaning and Mapping Yes/No Columns
Pertains to the following columns:
- 'Have you seen any of the 6 films in the Star Wars franchise?'
- 'Do you consider yourself to be a fan of the Star Wars film franchise?'

```python
yes_no = {'Yes': True, 'No': False}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
star_wars.head()
```

## Cleaning and Mapping Checkbox Columns

```python
import numpy as np

def cleaner(val):
    if val is np.nan:
        return False
    else:
        return True

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].apply(cleaner)
```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })
    
star_wars.head()
```

## Cleaning the Ranking Columns

```python
star_wars = star_wars[1:]
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })

star_wars.head()
```

## Finding the Highest-Ranked Movie

```python
star_wars[star_wars.columns[9:15]].mean()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

star_wars[star_wars.columns[9:15]].mean().plot.bar()
```

According to those who were surveyed, the original triology were all more liked than the prequel movies, with *Revenge of the Sith* being the least liked movie, and *The Empire Strikes Back* as the most liked (because lower ranking is better). This data shows that the general population is ignorant to the breathtaking action scenes of episodes 2 and 3.

## Finding the Most Viewed Movie

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
star_wars[star_wars.columns[3:9]].sum().plot.bar()
```

It appears that people have seen the original trilogy more than they have seen the prequels, with *The Empire Strikes Back* being the most watched episode, and *Revenge of the Sith* being the least watched episode. This makes sense, as the original movies should be watched before the prequels to have a better understanding of the Star Wars universe.

These results are very similar to the highest ranked movies, with episode 5 being the highest, and episode 3 being the lowest.

## Exploring the Data by Star Wars Fans & Non-fans

```python
star_wars_fan = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] == True]
star_wars_nonfan = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] == False]

star_wars_fan[star_wars_fan.columns[3:9]].sum().plot.bar()
plt.title('Fan Movies Seen')
plt.show()
star_wars_nonfan[star_wars_nonfan.columns[3:9]].sum().plot.bar()
plt.title('Non-fan Movies Seen')
plt.show()
```

It seems that fans of Star Wars have seen most of the movies at a fairly even amount, while non-fans have seen episodes 5 and 6 at a significantly higher frequency than episodes 2, 3, and 4.

```python
star_wars_fan[star_wars_fan.columns[9:15]].mean().plot.bar()
plt.title('Fan Movie Ranking')
plt.show()
star_wars_nonfan[star_wars_nonfan.columns[9:15]].mean().plot.bar()
plt.title('Non-fan Movie Rankings')
plt.show()
```

It appears that fans of Star Wars are more clear in likeing the original movies more than the prequels, while non-fans actually put episode 1 as the second most liked Star Wars movie. Fans also like episode 4 signifantly more than non-fans.
